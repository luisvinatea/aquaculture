```python
import pandas as pd
import urllib.parse

# File path for the direccion.csv
direccion_csv_filepath = '/home/luisvinatea/Data/Gdrive/BERAQUA/docs/datasets/csv/direccion.csv'

# Read the direccion CSV into pandas DataFrame
df_direccion = pd.read_csv(direccion_csv_filepath)

# Function to clean and format the address
def clean_and_format_address(row):
    # Combine direccion_1, direccion_2, direccion_3 while skipping 'nan' values
    address_parts = [str(row['direccion_1']), str(row['direccion_2']), str(row['direccion_3'])]
    # Remove parts that are 'nan' or empty strings
    clean_address = " ".join([part for part in address_parts if part and part.lower() != 'nan'])
    
    # Replace underscores with spaces for Google Maps compatibility
    clean_address = clean_address.replace('_', ' ')
    
    # If no valid address found, return None
    if not clean_address:
        return None
    
    # URL encode the address to safely include it in a Google Maps link
    return urllib.parse.quote_plus(clean_address)

# Apply the address formatting to each row
df_direccion['formatted_address'] = df_direccion.apply(clean_and_format_address, axis=1)

# Generate the Google Maps links
df_direccion['google_maps_link'] = 'https://www.google.com/maps?q=' + df_direccion['formatted_address']

# Optionally, save the DataFrame with Google Maps links to a new CSV file
output_csv_filepath = '/home/luisvinatea/Data/Gdrive/BERAQUA/docs/datasets/csv/direccion_with_links.csv'
df_direccion.to_csv(output_csv_filepath, index=False)

# Print the DataFrame with the new links to check
print(df_direccion[['Importer', 'google_maps_link']])
```
